```python
# Part A
# Apply smoothing filter on an image using convolution operation (eg. box filter)

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.tif')
kernel = np.ones((6,6),np.float32)/36
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Part B
# Apply sharpening filter on an image using convolution operation.

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.tif')
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Sharpening')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Part A 
# Add Gaussian noise to an image and apply average and Gaussian filters to the noisy image. Compare the output images.


import numpy as np
import matplotlib.pyplot as plt
import cv2

# Read the image
img = cv2.imread('lena.tif', 0)

# Add Gaussian noise
noise = np.random.normal(0, 50, img.shape)
noise_img = img + noise

# Apply average filter and Gaussian filter to noise_img
avg_img = cv2.blur(noise_img, (5, 5))
gauss_img = cv2.GaussianBlur(noise_img, (5, 5), 0)

# Display the images
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(noise_img, cmap='gray')
plt.title('Noisy image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(avg_img, cmap='gray')
plt.title('Average filtered image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(gauss_img, cmap='gray')
plt.title('Gaussian filtered image')
plt.axis('off')

plt.show()
```

```python
# Part B
# Add salt-and-pepper noise to an image and apply Gaussian and median filters to the noisy image. Compare the output images.

import numpy as np
import matplotlib.pyplot as plt
import cv2

# Read the image
img = cv2.imread('lena.tif', 0)

# Add salt and pepper noise
noise = np.random.randint(0, 2, img.shape)
noise_img = img.copy()
noise_img[noise == 0] = 0
# noise_img[noise == 1] = 255

# Apply Gaussian filter and median filter to noise_img

gauss_img = cv2.GaussianBlur(noise_img, (5, 5), 0)
median_img = cv2.medianBlur(noise_img, 5)

# Display the images
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(noise_img, cmap='gray')
plt.title('Noisy image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(gauss_img, cmap='gray')
plt.title('Gaussian filtered image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(median_img, cmap='gray')
plt.title('Median filtered image')
plt.axis('off')

plt.show()
```
